```python
ica = mne.preprocessing.ICA(n_components=20, random_state=97, max_iter=800)

ica.fit(raw.copy().filter(8, 35))
ica.plot_components()
```

```python
import numpy as np

# Calculate variance of each channel
channel_variances = np.var(raw.get_data(), axis=1)

# Calculate z-scores of these variances
z_scores = (channel_variances - np.mean(channel_variances)) / np.std(channel_variances)

# Consider channels with z-scores beyond an absolute value of 3 as bad
bad_channel_indices = np.where(np.abs(z_scores) > 3)[0]
bad_channel_names = [raw.ch_names[i] for i in bad_channel_indices]

print("Bad Channels:", bad_channel_names)

# Marking identified bad channels in the Raw object
raw.info['bads'] += bad_channel_names
```
